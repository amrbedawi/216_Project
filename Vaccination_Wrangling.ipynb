Now, condense the vaccination data into a few specific data points per country:
1. Vaccines per day (per capita) as of March 15, 2021.
2. Total vaccinated per capita as of March 15, 2021.

```python
mar14 = df[df['date']=='2021-03-14']
mar14.head()
mar14 = mar14[['country','people_fully_vaccinated_per_hundred','daily_vaccinations_per_million']]
```

```python
hap_vac = pd.merge(mar14, whr21_2020, left_on='country', right_on='Country name')
hap_vac
```

```python
hap_vac.dropna(subset=['people_fully_vaccinated_per_hundred'])
```

```python
sns.relplot(data=hap_vac.dropna(subset=['people_fully_vaccinated_per_hundred']), x='Freedom to make life choices', y='people_fully_vaccinated_per_hundred')
```
