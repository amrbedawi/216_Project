```python
import pandas as pd
import seaborn as sns
import numpy as np
sns.set_theme()
covid = pd.read_csv("owid-covid-data.csv")
#print(covid.shape)
#print(covid.head())
happiness= pd.read_csv("DataPanelWHR2021C2.csv")
country_cases=covid.groupby("location").sum()
last_year=happiness["year"]==2020
happiness_2020=happiness[last_year]
merged=pd.merge(country_cases, happiness_2020, how='left', on=['location'])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
merged=merged.dropna()
linear_model=LinearRegression()
linear_model.fit(X=merged[["Life Ladder"]].values, y=merged["new_cases_per_million"].values)
predicted_new_cases_per_million=linear_model.predict(merged[["Life Ladder"]].values)
print("MSE: ",mean_squared_error(merged["new_cases_per_million"].values, predicted_new_cases_per_million))
print("r^2: ", r2_score(merged["new_cases_per_million"].values, predicted_new_cases_per_million))
x=sns.regplot(data=merged, x="Life Ladder", y="new_cases_per_million")
x.set(xlabel ="Life Ladder", ylabel = "Total COVID Cases")
```
